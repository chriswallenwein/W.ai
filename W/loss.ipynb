Let's take a closer look at different loss and cost functions commonly used in machine learning and deep learning. In particular the L1 loss, L2 loss, Binary Cross Entropy Loss, Multiclass Cross Entropy Loss and the Hinge Loss.

Before we start with the actual functions, I want to make a distinction between the word loss and the word cost because they are often used interchangably. The loss references the penalty given for a single prediction while the cost references the penalty given for a set of multiple predictions (e.g. one batch or entire dataset). The cost if often computed by averaging the loss for each prediction

# L1 Loss
The L1 loss is defined as $L_1(y, \hat{y}) = \vert \hat{y} - y \vert$

For a fixed y = 0, the loss looks like this:

```python
# define y and y_hat
y = 0
y_hat = np.arange(-5, 6, 1)

# calculate loss
l1_loss = loss.L1()
L = l1_loss.forward(y, y_hat)

# plot result
fig = plt.figure()
ax = fig.add_subplot(111,aspect="equal")
plt.plot(y_hat, L)
plt.title("L1 loss for y = 0")
plt.xlabel("y_hat")
plt.xticks(np.arange(-5,6,1))
plt.ylabel("Loss")
plt.yticks(np.arange(0,6,1))
plt.show()
```

Just by looking at the plot, one can see that the derivative of the L1 loss is 1 for y_hat > y and -1 for y_hat < y. For y_hat = y the derivative is undefined. Let's set it to 0.

```python
l1_loss.forward(1,1)
```

# L2 Loss
The L2 loss is defined as $ L_2(y, \hat{y}) = (\hat{y} - y)^2 $
For a fixed y = 0, the loss looks like this:

```python
# define y and y_hat
y = 0
y_hat = np.arange(-3, 3.1, 0.1)

# calculate loss
l2_loss = loss.L2()
L = l2_loss.forward(y, y_hat)

# plot result
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(y_hat, L)
plt.title("L2 loss for y = 0")
plt.xlabel("y_hat")
plt.xticks(np.arange(-3,4,1))
plt.ylabel("Loss")
plt.yticks(np.arange(0,10,1))
plt.show()
```

Let's calculate the derivative of the L2 loss by applying the chain rule:

$L_2(y,\hat{y})=(\hat{y} - y)^2$

$\frac{\partial L}{\partial \hat{y}} = 2(\hat{y} - y) \cdot 1 = 2 (\hat{y} - y)$
